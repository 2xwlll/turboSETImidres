```python
#!/usr/bin/env python3

import pandas as pd

CSV = "/home/cgchoza/galaxies/complete_cadences_catalog.csv"

# MHz
TARGET_FREQS = [
    1604.0,
    1618.0,
    1623.0,
]

TOL = 2.0   # MHz window

df = pd.read_csv(CSV)

print("Columns:")
print(df.columns.tolist())
print()

matches = []

for freq in TARGET_FREQS:

    mask = (
        (df["f_start"] <= freq + TOL)
        &
        (df["f_stop"] >= freq - TOL)
    )

    found = df[mask].copy()
    found["target_freq"] = freq

    matches.append(found)

result = pd.concat(matches, ignore_index=True)

outfile = "lband_targets.csv"
result.to_csv(outfile, index=False)

print(f"Found {len(result)} matching rows")
print(f"Saved -> {outfile}")
```
